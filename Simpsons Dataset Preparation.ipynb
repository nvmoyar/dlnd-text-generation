```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv


data_path = 'data/simpsons/simpsons_script_lines.csv'

data = pd.read_csv(data_path, error_bad_lines=False)


data.head()
```

```python
df = pd.DataFrame(data, columns = ['raw_text', 'episode_id', 'number', 'location_id'])

df = df.sort_values(['episode_id', 'number'], ascending=True)

df.head()
```

```python
filter_by_episode = df['episode_id'] == 21
print(df[filter_by_episode ])

filter_by_location_id = df['location_id'] == 15
# print(df[filter_by_location_id])
```

```python
count = 1

outF = open("./data/simpsons/dataset.txt", "w")
    
for index, column in df.iterrows():
    
    if column['episode_id'] == count: 
    
        outF.write(column['raw_text'] + "\n")
    
    else: 
        
        outF.write("\n\n" + column['raw_text'] + "\n")
        count += 1

outF.close()
   
  
```

```python
df.to_csv('./data/simpsons/dataset.txt', mode = 'w', index=False, header=False, quoting=csv.QUOTE_MINIMAL, 
          columns=["raw_text", "episode_id"], sep='\t')
```
