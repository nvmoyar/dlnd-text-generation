```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv


data_path = 'data/simpsons/simpsons_script_lines.csv'

data = pd.read_csv(data_path, error_bad_lines=False)


data.head()
```

```python
df = pd.DataFrame(data, columns = ['raw_text', 'episode_id', 'number', 'location_id'])

df = df.sort_values(['episode_id', 'number'], ascending=True) #[:500]

print(df.shape)

df.head()
```

```python
filter_by_episode = df['episode_id'] == 2
print(df[filter_by_episode ])

filter_by_location_id = df['location_id'] == 15
# print(df[filter_by_location_id])
```

```python
count = 1

outF = open("./data/simpsons/dataset.txt", "w")
    
for index, column in df.iterrows():
       
    character_lines = column['raw_text'].split(":")
    character_lines[0] = character_lines[0].replace(" ", "_")
    character_lines[0] = character_lines[0].replace("(", "")
   
    
    if column['episode_id'] == count: 
                        
        outF.write(character_lines[0] + ": " + character_lines[1] + "\n")
    
    else: 
        
        outF.write("\n\n" + character_lines[0] + ": " + character_lines[1] + "\n")
        count += 1

outF.close()
   
  
```
